## **Linear Regression using Stats Models**

```python
# Stats model implementation for Linear Regression

import statsmodels.api as sn
X_train = sn.add_constant(X_train)
model = sn.OLS(y_train.values,X_train)
model = model.fit()
print(model.summary())
```

*   Adjusted. R-squared reflects the fit of the model. R-squared values range from 0 to 1, where a higher value
*   P >|t| is your p-value. A p-value of less than 0.05 is considered to be statistically significant.
*   Confidence Interval represents the range in which our coefficients are likely to fall (with a likelihood of 95%).


#### From above we see that for SOP the p value is very high hence this might be the main cause for multicollinearity in the model so lets remove the columns SOP & then lets calculate VIF and check.

## **Variance Inflation factor Calculation**

#### 1) VIF score of an independent variable represents how well the variable is explained by other independent
variables.
#### 2) So, the closer the R^2 value to 1, the higher the value of VIF and the higher the multicollinearity with the particular independent variable.


```python
# VIF Calculation
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['Features'] = X_train_new.columns
vif['VIF'] = [variance_inflation_factor(X_train_new.values,i) for i in range(X_train_new.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

#### Since none of the VIF values are greater than 10 we can presume that weare good now in terms of multicollinearity present between the features

## Bias-Variance Tradeoff

- **Bias:** Results from oversimplified model assumptions.
- **Variance:** Occurs when assumptions are too complex.
- The ideal model has low bias and low variance.

### Strategies to Address Bias and Variance:

- **Dimensionality Reduction and Feature Selection:**
  - Can decrease variance by simplifying models.

- **Training Set Size:**
  - Larger training sets tend to decrease variance.

### Reducing Bias:

- Change the model.
- Ensure the data is truly representative:
  - Diverse and represents all possible groups or outcomes.
- Parameter tuning.

### Bias-Variance Decomposition:

- Forms the conceptual basis for regression regularization methods (e.g., Lasso and Ridge Regression).
- Regularization introduces bias into the regression solution to reduce variance.
- Provides superior Mean Squared Error (MSE) performance compared to ordinary least squares (OLS) solutions.

**Note:**
Linear and Generalized Linear Models can be regularized to decrease variance at the cost of increasing bias.
